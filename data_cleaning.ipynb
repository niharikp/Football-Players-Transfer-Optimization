```python
games_2022 = games_df[games_df['year'] == 2022]

same_club_mask = games_2022['player_club_id'] == games_2022['player_current_club_id']

games_cur_club = games_2022[same_club_mask]

# Group by player_id and calculate the sum for each of the specified columns
player_stats_cur_club = games_cur_club.groupby('player_id')['goals', 'yellow_cards', 'red_cards', 'minutes_played', 'assists'].sum().reset_index()

# Rename columns for clarity if needed
player_stats_cur_club.rename(columns={'goals': 'total_goals_cur_club', 
                             'yellow_cards': 'total_yellow_cards_cur_club', 
                             'red_cards': 'total_red_cards_cur_club', 
                             'minutes_played': 'total_minutes_played_cur_club',
                             'assists': 'total_assists_cur_club'}, inplace=True)

player_stats_cur_club
```

```python
def calculate_performance_score(row):
    row['min_per_game'] = row['minutes_played']/row['games_played']
    if row['position'] == 'Attack':
        return ((row['goals'] * 0.3) + (row['assists'] * 0.25) -
                (row['yellow_cards'] * 0.05) - (row['red_cards'] * 0.1) +
                (row['minutes_played'] * 0.1) + (row['games_played'] * 0.1) +
                (row['min_per_game'] * 0.1))
    elif row['position'] == 'Midfield':
        return ((row['goals'] * 0.15) + (row['assists'] * 0.25) -
                (row['yellow_cards'] * 0.15) - (row['red_cards'] * 0.15) +
                (row['minutes_played'] * 0.1) + (row['games_played'] * 0.1) +
                (row['min_per_game'] * 0.1))
    elif row['position'] == 'Defender':
        return ((row['goals'] * 0.05) + (row['assists'] * 0.1) -
                (row['yellow_cards'] * 0.25) - (row['red_cards'] * 0.3) +
                (row['minutes_played'] * 0.1) + (row['games_played'] * 0.1) +
                (row['min_per_game'] * 0.1))
    else:
        return None
```

```python
# Define other weights and factors
EXPERIENCE_WEIGHT = 0.5
DECAY_2022 = 1.5
DECAY_2020 = 1.0
CAREER_WEIGHT = 0.5

# Function to calculate the overall performance score
def calculate_performance_score(row):
    experiece_factor = (row['year'] / row['age']) * EXPERIENCE_WEIGHT
    time_decay_factor = (row['performance_2022'] * DECAY_2022) + (row['performance_2020'] * DECAY_2020)
    #market_value_factor = (row['highest_market_value_in_eur'] - row['market_value_in_eur'])
    performance_factor = row['performance_career'] * CAREER_WEIGHT
    return experiece_factor + time_decay_factor + performance_factor

# Apply the performance score calculation
final_merged_df['raw_performance_score'] = final_merged_df.apply(calculate_performance_score, axis=1)

# Normalize the scores to a 0-50 scale
max_score = final_merged_df['raw_performance_score'].max()
final_merged_df['normalised_performance_score'] = (final_merged_df['raw_performance_score'] / max_score) * 50
```

```python
from gurobipy import Model, GRB, quicksum
import pandas as pd

# Load and prepare data
data_path = '/Users/niharikapatil/Documents/Academics/Fall Semester 2023/DABP/Final Project/data/final_merged.csv'  # Replace with actual path
players_data = pd.read_csv(data_path)

# Define constants and parameters
team_var = '131'  # Example team
budget_var = 150000000
attack_needed = 1
defender_needed = 1
midfield_needed = 1
CATEGORY_BUDGETS = {"Attack": 60000000, "Midfield": 40000000}

# Initialize the model
model = Model("FootballTeamTransfersOptimization")

# Decision variables
x = {}
for i in players_data['player_id']:
    x[(i, team_var)] = model.addVar(vtype=GRB.BINARY, name=f"x_{i}_{team_var}")

# Objective function
obj = quicksum(players_data.loc[players_data['player_id'] == i, 'market_value_in_eur'].iloc[0] *
               players_data.loc[players_data['player_id'] == i, 'normalised_performance_score'].iloc[0] * x[(i, team_var)]
               for i in players_data['player_id'])
model.setObjective(obj, GRB.MAXIMIZE)

# Constraints
# Budget constraint
model.addConstr(quicksum(players_data.loc[players_data['player_id'] == i, 'market_value_in_eur'].iloc[0] * x[(i, team_var)]
                         for i in players_data['player_id']) <= budget_var, name="TotalBudget")

# Positional need constraints
for k in ["Attack", "Midfield", "Defender"]:
    model.addConstr(quicksum(x[(i, team_var)] for i in players_data[players_data['position'] == k]['player_id']) >= NEEDS[team_var][k], name=f"Need_{k}")

# Categorical budget constraints
for k in CATEGORY_BUDGETS.keys():
    model.addConstr(
        quicksum(players_data.loc[(players_data['player_id'] == i) & (players_data['position'] == k), 'market_value_in_eur'].iloc[0] * x[(i, team_var)]
                 for i in players_data[players_data['position'] == k]['player_id']) <= CATEGORY_BUDGETS[k],
        name=f"CategoricalBudget_{k}"
    )

# One team per player constraint
for i in players_data['player_id']:
    model.addConstr(quicksum(x[(i, j)] for j in TEAMS) <= 1, name=f"OneTeam_{i}")

# Solve the model
model.optimize()

# Output results
team_player_dict = {}
for v in model.getVars():
    if v.x > 0:  # If the player is chosen in the solution
        _, player_id, team_id = v.varName.split('_')
        team_id = int(team_id)
        player_id = int(player_id)
        team_player_dict.setdefault(team_id, []).append(player_id)

selected_players = pd.DataFrame

for team, players in team_player_dict.items():
    print("Team:", team)
    selected_players = players_data[players_data['player_id'].isin(players)]
    print(selected_players[['player_id', 'name', 'position', 'normalised_performance_score', 'market_value_in_eur']])
```
